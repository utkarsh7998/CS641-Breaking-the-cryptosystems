```python
## This module does the differential cryptanalysis part on the 6th round
import numpy as np

#boutput file contains pairs of blocks of binary output
with open('binciphertexts.txt', 'r') as f:
    boutput = f.read().splitlines()

#consider 90000 binary cipher text values
boutput=boutput[0:90000]

# Storing Expansion Box, S-Box Input, S-Box Output XOR values in files
F_eop=open('eop.txt', 'w') #Expansion box output XOR
F_sip=open('sip.txt', 'w') #S-Box input XOR
F_sop=open('sop.txt', 'w') #S-Box output XOR

#Expansion Matrix
EBOX = [32, 1, 2, 3, 4, 5,
      4, 5, 6, 7, 8, 9,
      8, 9, 10, 11, 12, 13,
      12, 13, 14, 15, 16, 17,
      16, 17, 18, 19, 20, 21,
      20, 21, 22, 23, 24, 25,
      24, 25, 26, 27, 28, 29,
      28, 29, 30, 31, 32, 1]

#Final Inverse Permutation Matrix
FINVPER = [57,49,41,33,25,17,9,1,
          59,51,43,35,27,19,11,3,
          61,53,45,37,29,21,13,5,
          63,55,47,39,31,23,15,7,
          58,50,42,34,26,18,10,2,
          60,52,44,36,28,20,12,4,
          62,54,46,38,30,22,14,6,
          64,56,48,40,32,24,16,8]

#Inverse Permutation Matrix
INVPER = [9, 17, 23, 31,
       13, 28, 2, 18,
       24, 16, 30, 6,
       26, 20, 10, 1,
       8, 14, 25, 3,
       4, 29, 11, 19,
       32, 12, 22, 7,
       5, 27, 15, 21]

for i in range(0,len(boutput),2):
    op1 = boutput[i]
    op2 = boutput[i+1]
    invfp1 = []
    invfp2 = []
    sbox_input = []
    sbox_output = []
    expansion_ipr6 = [] #expansion input 6th round right half
    
    #Final Inverse Permutation of 1st and 2nd block
    for i in FINVPER:
        invfp1.append(op1[i-1])
        invfp2.append(op2[i-1])
    inversefp_xor=np.bitwise_xor(np.array(invfp1).astype(int),np.array(invfp2).astype(int))
    
    #Expansion Operation 
    for i in EBOX:
        sbox_input.append(inversefp_xor[i-1])
    F_sip.write(''.join(map(str,sbox_input))+"\n")
    inversefp_xor[37]=inversefp_xor[37]!=1
    
    #Inverse Permutation Operation
    for i in INVPER:
        sbox_output.append(inversefp_xor[32+i-1])
    F_sop.write(''.join(map(str,sbox_output))+"\n")
    
    #output of expansion for 1st block
    for i in EBOX:
        expansion_ipr6.append(invfp1[i-1])
    F_eop.write(''.join(map(str,expansion_ipr6))+"\n")
    
F_eop.close()
F_sip.close()
F_sop.close()
```
